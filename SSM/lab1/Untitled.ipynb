### 1. C помощью мультипликативного конгруэнтного метода (МКМ)

a_t = {b * a_{t-1} }Mod(M)


````python
def multiplicative_congruential_method (a_0, b, M, count):
    array = [a_0]
    for i in range(count):
        a_asterisk = b*array[i] % M
        array.append(a_asterisk)
    return array
````
